# **2.) Model Implementation**

```python
# Train-test-split
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.8,stratify=target,random_state=150)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Scaling
min_max=MinMaxScaler()
x_train=pd.DataFrame(min_max.fit_transform(x_train),columns=x_train.columns)
x_test=pd.DataFrame(min_max.transform(x_test),columns=x_test.columns)
```

```python
x_train
```

```python
x_test
```

```python
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()
```

```python
n_classes=26
epochs=20
batch_size=32 
```

```python
ann=tf.keras.models.Sequential()
```

```python
#To give no of units in hidden layer,we do a calculation ie;no of input neurons*2/3+no. of output neurons
784*(2/3)+10
16*2/3+1
```

```python
# so now we take roundly 12
```

```python
# Input layer,first hidden layer
ann.add(tf.keras.layers.Dense(units=12,activation='relu'))
```

```python
12*2/3+1
```

```python
# Second hidden layer
ann.add(tf.keras.layers.Dense(units=9,activation='relu'))
```

```python
# Output layer
ann.add(tf.keras.layers.Dense(units=n_classes,activation='softmax'))
```

```python
ann.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),loss=tf.keras.losses.sparse_categorical_crossentropy,metrics=['accuracy']) #Same loss function because ts same dataset for single and multi
```

```python
x_test.shape
```

```python
history1=ann.fit(x_train,y_train,epochs=epochs,batch_size=batch_size,validation_data=(x_test,y_test)) #Whenever you listen validation give testing data(Its my words)
```

```python
val_loss,val_acc=ann.evaluate(x_test,y_test) #Finding validation loss and acc so we given testing data if we want find last training accuracy and loss we can give training data simple!
print('val_loss:',val_loss,'val_acc:',val_acc)
```

# **3.) Hyperparameter Tuning**

```python
# Parameters
hidden_layers_list = [1,2,3]
neurons_list = [32,64]
activations = ['relu','tanh']
learning_rates = [0.001, 0.0005]
```

```python
results = []
for hidden_layers in hidden_layers_list:
    for neurons in neurons_list:
        for activation in activations:
            for lr in learning_rates:
                model = tf.keras.models.Sequential()
                # input + first hidden layer
                model.add(tf.keras.layers.Dense(neurons, activation=activation, input_shape=(16,)))
               
                # additional hidden layers
                for i in range(hidden_layers - 1):
                    model.add(tf.keras.layers.Dense(neurons, activation=activation))

                # output layer
                model.add(tf.keras.layers.Dense(26, activation='softmax'))

                optimizer = tf.keras.optimizers.Adam(learning_rate=lr)

                model.compile(optimizer=optimizer,loss='sparse_categorical_crossentropy',metrics=['accuracy'])

                history = model.fit(x_train,y_train,epochs=20,batch_size=32,validation_data=(x_test, y_test),verbose=0)

                val_acc = history.history['val_accuracy'][-1]

                results.append((hidden_layers, neurons, activation, lr, val_acc))

                print('No of hidden Layers:',hidden_layers,'Neurons:',neurons,'activation:',activation,'learning rate:',lr,'Accuracy:',val_acc)
```

```python
best = max(results, key=lambda x: x[4])
print("Best configuration:", best)
```

# **4.) Evaluation**

```python
y_pred = ann.predict(x_test)
y_pred = np.argmax(y_pred, axis=1)
```

```python
# Accuracy, Precision, Recall, and F1-score 
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

**Discussion of Performance Differences**

- The initial ANN model with **default hyperparameters** achieved an accuracy of **63.09%** on the test dataset.
- The performance was moderate because the model used default settings without optimizing the network architecture or learning parameters.
- After applying **hyperparameter tuning,** the model performance improved significantly.
- The tuned model achieved an accuracy of 90.03%, which is considerably higher than the default model.
- The best configuration included **3 hidden layers, 64 neurons per layer, tanh activation function, and a learning rate of 0.001,** which enhanced the model’s ability to learn complex patterns.
- Overall, **hyperparameter tuning played a crucial role in improving the classification performance of the ANN model.**
